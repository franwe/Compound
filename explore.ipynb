```python
import requests
from pprint import pprint
import pandas as pd 
```

```python
# The Graph - Current Borrows
  
url = "https://api.thegraph.com/subgraphs/name/graphprotocol/compound-v2"

# defining a params dict for the parameters to be sent to the API
query = """{
 markets {
    underlyingSymbol
    borrowRate
    totalBorrows
    collateralFactor
    underlyingPriceUSD
    blockTimestamp
    accrualBlockNumber
  }
}
"""

r = requests.post(url, json={'query': query})
data = r.json()["data"]["markets"]

graph_df = pd.DataFrame(data)
graph_df
```

```python
# Compound - Current Borrows - CTokenService
url = "https://api.compound.finance/api/v2/ctoken"

params = {
  "block_timestamp": 0, 
}

r = requests.get(url = url, params = params)
data = r.json()["cToken"]
```

```python
clean_data = []
for d in data:
    clean_d = {}
    for k, v in d.items():
        if type(v) == dict:
            clean_d[k] = v["value"]
        else:
            clean_d[k] = v
    clean_data.append(clean_d)

comp_df = pd.DataFrame(clean_data)
comp_df
```
